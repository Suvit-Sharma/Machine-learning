```python
carsales=pd.read_csv("carsales.csv")
carsales
```

```python
carsales["Sale Date"]=pd.date_range("1/1/2020",periods=10)
carsales
```

```python
carsales["Price"]=carsales["Price"].str.replace("[\$\,]","").astype(float)
carsales
```

```python
carsales["Total sale"]=carsales["Price"].cumsum()
carsales
```

```python
fig,ax=plt.subplots(1,2,figsize=(20,5))
ax[0].plot(carsales["Sale Date"],carsales["Total sale"])
ax[1].plot(carsales["Total sale"],carsales["Sale Date"]);
```

```python
over_50=heart[heart["age"]>50]
over_50.head()
```

```python
len(over_50)
```

```python
fig,axi=plt.subplots(figsize=(10,6))
ax=over_50.plot(kind="scatter",
               x="age",
               y="chol",
               c="target",
               ax=axi)
```

OO method from Scratch

```python
fig,ax=plt.subplots(figsize=(10,6))

#plotting data

scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
#customizing plot 

ax.set(title="heart disease vs age",
            xlabel="age",
            ylabel="chol");

#creating legend

ax.legend(*scatter.legend_elements(),title="target");

#adding a horizontal line

ax.axhline(over_50["chol"].mean(),
          linestyle=("--"));
```

making same graph but on subplots

```python
# Making a plot

fig,ax = plt.subplots(2,1,figsize=(10,10),
                     sharex=True)

# Adding data

scatter = ax[0].scatter(x=over_50["age"],
                         y=over_50["chol"],
                         c=over_50["target"]);

# Customizing the plot 
ax[0].set(title="heart disease vs chol",
      ylabel="chol");

# Making a horizontal line'

ax[0].axhline(over_50["chol"].mean(),linestyle=("--"))

# Making legend 

ax[0].legend(*scatter.legend_elements(),title="target");

# Making plot between heart disease and thalach

scatter=ax[1].scatter(x=over_50["age"],
                 y=over_50["thalach"],
                 c=over_50["target"]);
# Making legend

ax[1].legend(*scatter.legend_elements(),title="target");

# Customizing plot 

ax[1].set(title="heart disease vs max heart rate(Thalach)",
         xlabel="age",
         ylabel="max heart rate(thalach)");

# Making a mean line 

ax[1].axhline(over_50["thalach"].mean(),linestyle="--");

# Making figure heading

fig.suptitle("heart disease vs thalach and cholestrol",fontsize=16,fontweight="bold")
```
